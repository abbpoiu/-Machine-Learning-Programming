```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 데이터 준비
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Raisin_Dataset.xlsx")
df  # print(df) 와 동일
```

```python
#칼럼명 확인을 해야됨.(label) -> 여기서는 Class가 라벨인듯
df.columns
```

```python
#칼럼명 확인 후 결측지 확인을 해야됨
#이 데이터는 결측치가 없으니 결측치 제거 X 따로 df.dropna()를 안해도됨.
df.isnull().sum()
```

```python
#레이블 불균형 확인
#어디에 초점이 맞춰져있는 데이터인지 확인.
df['Class'].value_counts()
```

```python
#레이블 갯수확인
#결과를 확인해보니 균형 데이터임.
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="Class")
plt.xlabel("Class")   # 건포도의 종류
plt.ylabel("Count")
plt.title("Label distribution")
plt.show()
```

```python
#날짜나 ID같은 불필요 컬럼을 찾지 못함. 칼럼 삭제 X
# 모든 레이블을 숫자로 변형

from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df
```

```python
# 입력과 출력 분리

X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X
```

```python
y
```

```python
# 훈련과 테스트 데이터 분리


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42

)
```

```python
# 각 데이터의 shape 확인 (전체데이터 수, 컬럼숫자)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#knn 모델을 위한 스케일링.
#X, y 데이터를 분리 후 X 데이터에 스케일링을 적용한 것.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)
```

```python
# -----------------------------
# 2) 모델 구성
# -----------------------------
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=500)
```

```python
#KNN 모델 구성
from sklearn.neighbors import KNeighborsClassifier

#스케일러가 필요한 이유
#데이터에서 큰값과 0~1근처의 값이 있는 컬럼이 동시에 존재할 경우
#knn은 거리기반 모델이라 큰값에 의해 의사결정이 왜곡될 가능성이 높음.
#근데 이제 스케일러의 위치가 잘못됐다. X와 y로 나눈 이후 바로 X에 스케일러를 쓰는게 맞음.

knn_clf = KNeighborsClassifier(
    n_neighbors=5,
    weights="distance",
    metric="minkowski",
    p=2
)
```

```python
#KNN 사용 (학습/평가)

knn_clf.fit(X_train_s, y_train)
knn_y_pred = knn_clf.predict(X_test_s)
```

```python
# -----------------------------
# 3) 모델 학습
# -----------------------------

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
```

```python
# -----------------------------
# 4) 예측결과 생성
# -----------------------------


dt_y_pred = dt.predict(X_test)
rf_y_pred = rf.predict(X_test)
lr_y_pred = lr.predict(X_test)
```

```python
# -----------------------------
# 5) 예측결과와 정답을 비교한 정확도 평가
# -----------------------------

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

dt_acc = accuracy_score(y_test, dt_y_pred)
rf_acc = accuracy_score(y_test, rf_y_pred )
lr_acc = accuracy_score(y_test, lr_y_pred)


print("=== DT Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print (confusion_matrix(y_test,dt_y_pred))

print("=== RF Test Accuracy ===")
print(f"Random Forest : {rf_acc:.4f}")
print (confusion_matrix(y_test,rf_y_pred))

print("=== LR Test Accuracy ===")
print(f"Logistic Reg. : {lr_acc:.4f}")
print (confusion_matrix(y_test,lr_y_pred))
```

```python
#KNN 정확도 평가
from sklearn.metrics import accuracy_score, confusion_matrix

knn_acc = accuracy_score(y_test, knn_y_pred)

print("=== KNN Test Accuracy ===")
print(f"KNN : {knn_acc:.4f}")
print(confusion_matrix(y_test, knn_y_pred))
```
